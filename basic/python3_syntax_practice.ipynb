## 3. 容器类型与切片

- 列表 `list`：可变序列，对应 JS 的 Array
- 元组 `tuple`：不可变序列，可用于多值返回
- 字典 `dict`：键值映射，对应 JS 的 Object 或 Map
- 集合 `set`：元素唯一、无序
- 切片 `sequence[start:end:step]` 获取子序列


```python
nums = [1, 2, 3, 4, 5]
first = nums[0]
sub = nums[1:4]
step = nums[::2]

point = (3, 4)
x_coord, y_coord = point

person = {"name": "Alice", "age": 30}
name = person["name"]
keys = list(person.keys())

unique = {1, 2, 2, 3}
first, sub, step, x_coord, y_coord, name, keys, unique
```

## 4. 控制流与循环

- `if`、`elif`、`else` 分支
- `for` 遍历可迭代对象，如列表、字符串、字典
- `range(n)` 生成整数序列
- `while` 循环适合不确定次数的重复


```python
n = 7
if n % 2 == 0:
    result = "even"
elif n % 3 == 0:
    result = "divisible by 3"
else:
    result = "other"

items = ["a", "b", "c"]
upper_items = []
for item in items:
    upper_items.append(item.upper())

total = 0
i = 1
while i <= 5:
    total += i
    i += 1

result, upper_items, total
```

## 5. 函数与参数

- 使用 `def` 定义函数
- 函数可以有默认参数、可变参数 `*args`、关键字参数 `**kwargs`
- Python 中函数是一等公民，可以作为参数传递


```python
def add(a, b):
    return a + b

def greet(name, prefix="Hello"):
    return prefix + ", " + name

def total_length(*texts):
    length = 0
    for t in texts:
        length += len(t)
    return length

add(2, 3), greet("Python"), total_length("ab", "cde", "f")
```

## 6. 类与面向对象

- 使用 `class` 定义类
- `__init__` 方法在实例创建时调用
- `self` 表示实例本身，类似 Java 中的 `this`
- 简单了解属性与方法的定义与访问


```python
class Counter:
    def __init__(self, start=0):
        self.value = start

    def increment(self):
        self.value += 1

    def reset(self):
        self.value = 0

c = Counter()
c.increment()
c.increment()
current = c.value
c.reset()
current, c.value
```

## 7. 推导式与迭代

- 列表推导式可以用更紧凑的方式生成列表
- 字典推导式、集合推导式语法类似
- 生成器表达式惰性计算，适合处理大数据流


```python
squares = [n * n for n in range(6)]
even_squares = [n for n in squares if n % 2 == 0]

name_lengths = {name: len(name) for name in ["Alice", "Bob", "Charlie"]}

gen = (n * n for n in range(3))
values = list(gen)

squares, even_squares, name_lengths, values
```
